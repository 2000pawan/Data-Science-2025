# K-Means Clustering

# Income Data

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples,silhouette_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import warnings
```

```python
warnings.filterwarnings('ignore')
df=pd.read_csv('G:/dataset/MLbookdataset/ML/Ml/32. K-Means clustering/income.csv')
plt.scatter(df['Age'],df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.show()
model=KMeans(n_clusters=3,n_init='auto',random_state=4)
y_pred=model.fit_predict(df[['Age','Income($)']])
print(y_pred)
df['cluster']=y_pred
print(model.cluster_centers_)
X=model.cluster_centers_[:,0]
y=model.cluster_centers_[:,1]
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color='green',marker='o')
plt.scatter(df2.Age,df2['Income($)'],color='red',marker='d')
plt.scatter(df3.Age,df3['Income($)'],color='blue',marker='s')
plt.scatter(X,y,color='purple',marker='*',label='centroid')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
plt.show()
sc=MinMaxScaler()
sc.fit(df[['Income($)']])
df['Income($)']=sc.transform(df[['Income($)']])
sc.fit(df[['Age']])
df['Age']=sc.transform(df[['Age']])
wsse=[]
for i in range(1,10):
    model=KMeans(n_clusters=i)
    model.fit(df[['Age','Income($)']])
    wsse.append(model.inertia_)
print(wsse)
plt.xlabel('Range(1,10)')
plt.ylabel('Sum of Squared Error(WSSE)')
plt.plot(range(1,10),wsse)
plt.show()
x=df[['Age','Income($)']]
Y=model.labels_
print(Y)
print('Silhouette_samples:-  ',silhouette_samples(x,Y))
print('Silhouette_Score is:-  ',silhouette_score(x,Y))
```

# Iris Data

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples,silhouette_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv('G:/dataset/classification/iris.csv')
df.drop(['Unnamed: 0','target'],axis='columns',inplace=True)
sc=MinMaxScaler()
sc.fit(df[['petal length (cm)']])
df['petal length (cm)']=sc.transform(df[['petal length (cm)']])
sc.fit(df[['petal width (cm)']])
df['petal width (cm)']=sc.transform(df[['petal width (cm)']])
X=df.iloc[:,2:]
wsse=[]
for i in range(1,10):
    model=KMeans(n_clusters=i)
    model.fit(X)
    wsse.append(model.inertia_)
print(wsse)
plt.xlabel('Range(1,10)')
plt.ylabel('Sum of Squared Error(WSSE)')
plt.plot(range(1,10),wsse)
plt.show()
model=KMeans(n_clusters=3)
y_pred=model.fit_predict(X)
print(y_pred)
df['cluster']=y_pred
print(model.cluster_centers_)
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='green',marker='o')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='red',marker='d')
plt.scatter(df3['petal length (cm)'],df3['petal width (cm)'],color='blue',marker='s')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.legend()
plt.show()
df.head()
Y=model.labels_
print(Y)
print('Silhouette_samples:-  ',silhouette_samples(X,Y))
print('Silhouette_Score is:-  ',silhouette_score(X,Y))
```
