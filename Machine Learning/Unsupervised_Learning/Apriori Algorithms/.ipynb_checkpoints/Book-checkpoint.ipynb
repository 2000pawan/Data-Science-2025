# Apriori Algorithm

# Market Basket Analysis

```python
from apyori import apriori
```

```python
file=open('G:/dataset/MLbookdataset/ML/Ml/33. Apriori/Market_Basket_Optimisation.csv','r')
x=[]
for i in file:
    i=i.split(',')
    x.append(i)
# print(x)
model=apriori(x,min_support = 0.003,min_confidence=0.4,min_lift=4,min_length=2)
f1=open('results_market_basket_analysis.txt','w')
for i in model:
    f1.write(str(i)+'\n')
f1.close()
```

```python
file=open('G:/dataset/MLbookdataset/ML/Ml/33. Apriori/Market_Basket_Optimisation.csv','r')
x=[]
for i in file:
    i=i.split(',')
    x.append(i)
# print(x)
model=apriori(x,min_support = 0.003,min_confidence=0.4,min_lift=4,min_length=2)
results=list(model)   
for item in results:  
    pair = item[0]   
    items = [x for x in pair]  
    print("Rule: " + items[0] + " -> " + items[1])  
  
    print("Support: " + str(item[1]))  
    print("Confidence: " + str(item[2][0][2]))  
    print("Lift: " + str(item[2][0][3]))  
    print("=====================================")  
```
