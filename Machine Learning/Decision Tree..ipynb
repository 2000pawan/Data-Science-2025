```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier,plot_tree
import matplotlib.pyplot as plt
```

## Gini finding formula
- gini(S)= 1-sum of all (p(each class)^2)
- weighted average gini= (left_node_gini * total sample of left_node/total sample of parent class)+(right_node_gini * total sample of right_node/total sample of parent class)
- threshold finding -- first find the importance feature using anova test then create a set of that feature and assume each value and range of min and max value of that feature and find the weighted average gini and min of that gini is the threshold value of that feature.

```python
df=load_iris()
X=df.data
y=df.target
```

```python
# importance hyperparameter using in decision tree
model=DecisionTreeClassifier(criterion='gini',random_state=10,max_depth=None,max_leaf_nodes=None,max_features=None,min_samples_split=2)
model.fit(X,y)
```

```python
plt.figure(figsize=(10,10))
plot_tree(model,feature_names=['SL','SW','PL','PW'],fontsize=8,filled=True)
plt.show()
```

```python
model.feature_importances_
```
