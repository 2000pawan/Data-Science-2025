## Ensemble Learning
- Ensemble Learning is a technique that use to train model for classification task. it's group of algorithms different type of algorithm that build multiple weak model and combined them to build a conceptual strong model to predict.
- 1. Bagging --> BaggimgClassifier, RandomForestClassifier etc. 
  2. Boosting --> AdaBoost, XGBoost, GradientBoost etc.
  3. Stacking --> VotingClassifier, StackingClassifier etc.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
```

```python
df=load_iris()
X=df.data
y=df.target
```

```python
model=RandomForestClassifier(n_estimators=200,random_state=10,bootstrap=True,oob_score=True,max_features='sqrt')
model.fit(X,y)
```

```python
sample=[[0.3,3.0,2.6,0.4]]
model.predict(sample)
```

```python
model.predict_proba(sample)
```

```python
model.predict_log_proba(sample)
```

```python
model.estimator_
```

```python
model.classes_
```

```python
model.estimator_params
```

```python
model.feature_importances_
```

```python
model.oob_score_
```

```python
model.score(X,y)
```

```python
from sklearn.ensemble import BaggingClassifier,VotingClassifier,AdaBoostClassifier
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

## Bagging Classifier same as Random Forest Algorithm but main difference is in Bagging Classifier we have to option to choose different classification algorithm unlike decision tree that use by random forest in random forest we only use decision tree as a week model. but in Bagging classifier we use different algorithm like Logistic regression, decision tree, svc, naive bayes etc if we not tell algorithm by default it is use decision tree.

```python
df=load_iris()
X=df.data
y=df.target
```

```python
'''all hyperparameter are same as random forest'''

weak_model=LogisticRegression(max_iter=200)
model=BaggingClassifier(estimator=weak_model,n_estimators=100)
model.fit(X,y)
```

## Stacking --> Voting Classifier is used same as bagging algorithm but main difference is in voting classifier and stacking we have to option to train weak model using multiple different classification algorithm that not provided by bagging 

```python
df=load_iris()
X=df.data
y=df.target
```

```python
wk_model=[('lr',LogisticRegression(max_iter=200)),('svc',SVC())]
model=VotingClassifier(estimators=wk_model)
model.fit(X,y)
```

## Boosting Algorithm --> 1. AdaBoost algorithm is used boosting technique to train weak model that fully different from bagging and stacking technique in boosting we first assign sample weight for all sample and then train model and after training calculate the model weight. after train first weak model its update sample weight according to model performance for e.g in a dataset total sample is 50 and model predict after training 45 correct and 5 wrong then correct sample weight is decrease and wrong sample weight is increase and then train new weak model and calculate new weak model weight continue repeat until the max estimatior reached or model classified all sample correct. in boosting weak model train sequentially one after one but in bagging and stacking weak model train parallel. new weight assign for first weak model is using this formula 1/total no of sample. after first model new weight update using 

## new weight increase=previous weight * exp(previous model weight)
## new weight decrease=previous weight * exp(-previous model weight)
## model_weight=1/2loge(1-TE/TE)  # TE=total error in model(sum of weights of misclassified sample)

```python
df=load_iris()
X=df.data
y=df.target
```

```python
wk_model=LogisticRegression(max_iter=200)
model=AdaBoostClassifier(estimator=wk_model,n_estimators=100,algorithm='SAMME')
model.fit(X,y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
wk1_model=KNeighborsClassifier()
model=AdaBoostClassifier(estimator=wk1_model,n_estimators=100,algorithm='SAMME')
model.fit(X,y)
```

```python
weak_model=SVC()
model=AdaBoostClassifier(estimator=wk_model,n_estimators=100,algorithm='SAMME')
model.fit(X,y)
```

```python
weak_model=SVC()
model=AdaBoostClassifier(estimator=wk_model,n_estimators=100,algorithm='SAMME.R')
model.fit(X,y)
```
