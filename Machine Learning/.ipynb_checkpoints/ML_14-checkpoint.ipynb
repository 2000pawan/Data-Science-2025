> Types of Loss/error in Regression.
------------------------------------------>
1. Residual--------------------------------> error in individual sample.
2. OLS(Ordinary Least Squared Error)-------> Sum of squares residual.
3. MSE(Mean Squared Error)-----------------> sum of squares residual/total no. of samples.
4. RMSE(Root Mean Squared Error)-----------> sqrt(MSE)
5. MAE(Mean Absolute Error)----------------> mean(abs(residuals))

etc.......

```python
residuals=y-pred
print(residuals)
OLS=sum(residuals**2)
print(OLS)
MSE=OLS/len(y)
print(MSE)
RMSE=math.sqrt(MSE)
print(RMSE)
MAE=sum(abs(residuals))/len(y)
print(MAE)
```

```python
MAE=abs(residuals).mean()
print(MAE)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error
```

```python
print(mean_absolute_error(y,pred))
print(mean_squared_error(y,pred))
print(root_mean_squared_error(y,pred))
```
