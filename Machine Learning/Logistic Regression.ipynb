## maths behind LogisticRegression
- first it builds a linear model to find coefs and intercept
- then convert this linear model to non linear sothat it can perform classification.
- logit fun is used to convert linear part to non linear
- sigmoid & softmax are used as logit funs.
- with binary classification,sigmoid is used and with multiclass softmax is used.

## Binary Classification 

```python
df=pd.read_csv("G:/dataset/classification/fruits.csv")
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
model=LogisticRegression()
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

## Linear Part using code

```python
sample=[3.5,40]
z=-0.98517568*3.5+-0.02200477*40+5.91844611
print(z)
```

```python
print(model.decision_function([sample]))
```

## Non Linear Part using Sigmoid function (Logit Function)
- sigmoid func= 1/1+e^-z   ## this formula gives you probability of class 1 or +ve class.
- if you want to determine the probability of class 0 then
- sigmoid function= 1/1+e^z 

```python
# Probability of class 1
result=1/(1+np.exp(-z))
print(result)
if result<=.5:
    print("Apple")
else:
    print("Banana")
```

```python
print(model.predict([sample]))
print(model.predict_proba([sample]))
```

```python
# probability of class 0 
result=1/(1+np.exp(z))
print(result)
```

```python
# also you try 1-probabilitty of class 1
p_0=1-1/(1+np.exp(-z))
p_0
```

## Multi Class Classification

## Multiclass classification used Softmax function to convert linear part into non linear part.
- there are two approach.
- 1. OVR (One vs Rest) --> in this we train multi model according to how many class in your dataset in this we let update +ve class in each model. Then used sigmoid to solve the multiclass problem.
- 2. Multinomial (softmax) --> in this we train each class separate model and the convert each model to non linear part and max of the probablity is our output

```python
from sklearn.datasets import load_iris
```

## OVR 

```python
df=load_iris()
X=df.data
y=df.target

model=LogisticRegression(multi_class='ovr',max_iter=200)
model.fit(X,y)
```

```python
sample=[1.5,.8,3.5,.9]
print(model.predict([sample]))
print(model.predict_proba([sample]))
```

```python
model.coef_
```

```python
model.intercept_
```

```python
z0=-0.44565052*1.5+  0.90054581*.8+ -2.32023343*3.5+ -0.98112874*.9+6.68931554
z1=-0.17917535*1.5+ -2.12878487*.8 + 0.69671839*3.5+ -1.27500517*.9+ 5.58592907
z2=-0.39270716*1.5+ -0.51442952*.8+  2.93064063*3.5+  2.41608802*.9+-14.43598319
print([z0,z1,z2])
```

```python
model.decision_function([sample])
```

```python
p0=1/(1+np.exp(-z0))
p1=1/(1+np.exp(-z1))
p2=1/(1+np.exp(-z2))

print([p0,p1,p2])
```

```python
# Normalized probability
pn0=p0/(p0+p1+p2)
pn1=p1/(p0+p1+p2)
pn2=p2/(p0+p1+p2)

print([pn0,pn1,pn2])
```

```python
print(model.predict_proba([sample]))
```

## Multinomial

```python
df=load_iris()
X=df.data
y=df.target

model=LogisticRegression(multi_class='multinomial',max_iter=200)
model.fit(X,y)
```

```python
sample=[1.5,.8,3.5,.9]
print(model.predict([sample]))
print(model.predict_proba([sample]))
```

```python
model.coef_
```

```python
model.intercept_
```

```python
z0=-0.41962017*1.5+  0.96690033*.8+ -2.51934621*3.5+ -1.08448699*.9+  9.83859123
z1=0.53107361*1.5+ -0.31498782*.8+ -0.20238066*3.5+ -0.94412374*.9+2.22306076
z2=-0.11145345*1.5+ -0.65191251*.8+  2.72172688*3.5+  2.02861073*.9+-12.061652 

print([z0,z1,z2])
```

```python
model.decision_function([sample])
```

```python
p0=np.exp(z0)/(np.exp(z0)+np.exp(z1)+np.exp(z2))
p1=np.exp(z1)/(np.exp(z0)+np.exp(z1)+np.exp(z2))
p2=np.exp(z2)/(np.exp(z0)+np.exp(z1)+np.exp(z2))

print([p0,p1,p2])
```

```python
model.predict_proba([sample])
```
