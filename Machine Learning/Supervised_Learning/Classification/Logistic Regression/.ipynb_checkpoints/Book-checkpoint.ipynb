# Logistic Regression

# Insurance Data

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df=pd.read_csv('G:/dataset/MLbookdataset/ML/Ml/26. Logistic Regression/insurance_data.csv')
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
plt.xlabel('Age')
plt.ylabel('Have Insurance')
plt.scatter(X,y,color='red',marker='+')
plt.show()
sn.regplot(x='age',y='bought_insurance',data=df,logistic=True,marker='+',color='blue')
plt.show()
model=LogisticRegression()
model.fit(X,y)
print('Accuracy Score is:- ',model.score(X,y))
```

```python
model.predict([[25],[56],[36]])
```

# HR data

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.model_selection import train_test_split
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
df1=pd.read_csv('G:/dataset/MLbookdataset/ML/Ml/26. Logistic Regression/HR_comma_sep.csv')
df=pd.read_csv('G:/dataset/MLbookdataset/ML/Ml/26. Logistic Regression/HR_comma_sep.csv')
left=df[df.left==1]
print(left.shape)
retained=df[df.left==0]
print(retained.shape)
avg=df.groupby('left').mean()
print(avg)
pd.crosstab(df.dept,df.left).plot(kind='bar')
plt.show()
df.drop(columns=['last_evaluation','number_project','time_spend_company','Work_accident','left','dept'],axis=1,inplace=True)
salary_dummy=pd.get_dummies(df['salary'],prefix='salary')
df=pd.concat([df,salary_dummy],axis='columns')
df.drop(columns=['salary','salary_medium'],axis=1,inplace=True)
X=df
y=df1.left
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
model=LogisticRegression()
model.fit(x_train,y_train)
print('Accuracy Score Train data is:- ',model.score(x_train,y_train))
print('Accuracy Score Test data is:- ',model.score(x_test,y_test))
```

```python
# 1 means leave company 0 means stay in company 
inputs=[[0.11,286,0,0,0]]
print(model.predict(inputs))
inputs=[[0.48,228,0,0,1]] 
print(model.predict(inputs))
```

# Digit Data

```python
from sklearn.datasets import load_digits
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df=load_digits()
print(dir(df))
print(df.data.shape)
# print(df.data)
# print(df.target)
# print(df.images)
print(df.data[8])
plt.gray()
plt.matshow(df.images[8])
plt.show()
print(df.target[8])
x_train,x_test,y_train,y_test=train_test_split(df.data,df.target,test_size=0.3)
model=LogisticRegression()
model.fit(x_train,y_train)
print('Accuracy Score Train data is:- ',model.score(x_train,y_train))
print('Accuracy Score Test data is:- ',model.score(x_test,y_test))
```

```python
plt.matshow(df.images[67])
plt.show()
print(model.predict([df.data[67]]))
print(df.target[67])
```

```python
for i in range(30,35):
    plt.matshow(df.images[i])
    plt.show()
print(model.predict(df.data[30:35]))
print(df.target[30:35])
```
