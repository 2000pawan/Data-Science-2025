# Simple Linear Regression

# Home Price Data

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv("G:/dataset/MLbookdataset/ML/Ml/19. Simple Linear Regression/homeprices.csv")
print(df)
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
sn.scatterplot(data=df,x='area',y='price')
plt.show()
model=LinearRegression()
model.fit(X,y)
print('Coefficient:-',model.coef_)
print('Intercept:-',model.intercept_)
print(model.predict([[3300]]))
y_predict=model.predict(X)
print('Accuracy Score r2_score:- ',r2_score(y,y_predict))
sn.lmplot(data=df,x='area',y='price')
plt.show()
```

# Salary Data

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("G:/dataset/MLbookdataset/ML/Ml/19. Simple Linear Regression/Salary_Data.csv")
print(df)
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
plt.scatter(X,y,color='red')
plt.show()
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=10)
model=LinearRegression()
model.fit(X_train,y_train)
y_predict=model.predict(X_train)
y_predict_test=model.predict(X_test)
print('Accuracy Score r2_score Train Sample:- ',r2_score(y_train,y_predict))
print('Accuracy Score r2_score Test Sample:- ',r2_score(y_test,y_predict_test))
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,y_predict,color='blue')
plt.title('Experience VS Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

# Multiple Linear Regression

# HomePrices

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
import math
```

```python
df=pd.read_csv("G:/dataset/MLbookdataset/ML/Ml/20. Multiple Linear Regression/homeprices.csv")
print(df)
print('-----------------------------------------------------------------------------------')
print('-----------------------------------------------------------------------------------')
df.isnull().sum()
med=math.floor(df.bedrooms.median())
df.bedrooms=df.bedrooms.fillna(med)
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
model=LinearRegression()
model.fit(X,y)
y_pred=model.predict(X)
print('-----------------------------------------------------------------------------------')
print('-----------------------------------------------------------------------------------')
print('Coeficent:- ',model.coef_)
print('Intercept:- ',model.intercept_)
print('Accuracy Score is :- ',model.score(X,y))
print('Accuracy Score is :- ',r2_score(y,y_pred))
print(model.predict([[3000,3,40]]))
print(model.predict([[2500,4,5]]))
print('-----------------------------------------------------------------------------------')
print('-----------------------------------------------------------------------------------')
sn.lmplot(x='area',y='price',data=df)
sn.lmplot(x='bedrooms',y='price',data=df)
sn.lmplot(x='age',y='price',data=df)
plt.show()
```

# Homes Excel

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
df=pd.read_excel("G:/dataset/MLbookdataset/ML/Ml/20. Multiple Linear Regression/ca11-03homes.xls")
print(df)
df.isnull().sum()

print('-----------------------------------------------------------------------------------')
print('-----------------------------------------------------------------------------------')
df.drop(columns=['Garage','Zip','Obs'],inplace=True)
print(df)

print('-----------------------------------------------------------------------------------')
print('-----------------------------------------------------------------------------------')
q3=df['Price'].quantile(0.75)
print('Q3:- ',q3)
q1=df['Price'].quantile(0.25)
print('Q1:- ',q1)
iqr=q3-q1
print('IQR:- ',iqr)
ul=q3+(1.5*iqr)
ll=q1-(1.5*iqr)
print('Upper Limit:- ',ul)
print('Lower Limit:- ',ll)
upper=np.where(df['Price']>=ul)
lower=np.where(df['Price']<=ll)
df.drop(upper[0],inplace=True)
df.drop(lower[0],inplace=True)

print('-----------------------------------------------------------------------------------')
print('-----------------------------------------------------------------------------------')
X=df.iloc[:,1:].values
y=df.iloc[:,0].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_train)
y_predict=model.predict(X_test)
print('Coeficent:- ',model.coef_)
print('Intercept:- ',model.intercept_)
print('Accuracy Score Training dataset is :- ',r2_score(y_train,y_pred))
print('Accuracy Score Testing dataset is :- ',r2_score(y_test,y_predict))
print(model.predict([[780,3,1]]))
print(model.predict([[1500,3,2],[2000,4,4]]))

print('-----------------------------------------------------------------------------------')
print('-----------------------------------------------------------------------------------')
sn.lmplot(x='SqFt',y='Price',data=df)
sn.lmplot(x='BedRooms',y='Price',data=df)
sn.lmplot(x='Baths',y='Price',data=df)
plt.show()
```

# Salary_Position 

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
```

```python
# Without adding Polynomial Features
df=pd.read_csv('G:/dataset/MLbookdataset/Ml/Ml/22. Polynomial Linear Regression/Salary_Position.csv')
X=df.iloc[:,1:2].values
y=df.iloc[:,-1].values
model=LinearRegression()
model.fit(X,y)
y_pred=model.predict(X)
print('Accuracy Score is:- ',r2_score(y,y_pred))
plt.scatter(X,y,color='red')
plt.show()
plt.scatter(X,y,color='red')
plt.plot(X,y_pred,color='blue')
plt.show()
```

```python
# With adding Polynomial Features
for i in range(1,10):
    pl=PolynomialFeatures(degree=i)
    X1=pl.fit_transform(X)
    model=LinearRegression()
    model.fit(X1,y)
    y_pre=model.predict(X1)
    r2=r2_score(y,y_pre)
    plt.scatter(X,y,color='red')
    plt.plot(X,y_pre,color='blue')
    plt.title(f'Degree {i}\nAccuracy Score:- {r2}',color='red')
    plt.show()
    print('---------------------------------------------------------------------------------------------------------------------')
    print('---------------------------------------------------------------------------------------------------------------------')
```

```python
pf=PolynomialFeatures(degree=7)
X2=pf.fit_transform(X)
model=LinearRegression()
model.fit(X2,y)
y_pr=model.predict(X2)
print('Accuracy Score Without adding Polynomial Features is:- ',r2_score(y,y_pred))
print('Accuracy Score With adding Polynomial Features is:- ',r2_score(y,y_pr))
```

# Salary_Experience

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
```

```python
# Without adding Polynomial Features
df=pd.read_csv('G:/dataset/MLbookdataset/Ml/Ml/22. Polynomial Linear Regression/Salary_Experience.csv')
X=df.iloc[:,0:1].values
y=df.iloc[:,-1].values
model=LinearRegression()
model.fit(X,y)
y_pred=model.predict(X)
print('Accuracy Score is:- ',r2_score(y,y_pred))
plt.scatter(X,y,color='red')
plt.xlabel('Years of Experience')
plt.ylabel("Salary")
plt.title('Salary v/s Experience')
plt.show()
```

```python
# With adding Polynomial Features
for i in range(1,10):
    pl=PolynomialFeatures(degree=i)
    X1=pl.fit_transform(X)
    model=LinearRegression()
    model.fit(X1,y)
    y_pre=model.predict(X1)
    r2=r2_score(y,y_pre)
    plt.scatter(X,y,color='red')
    plt.plot(X,y_pre,color='blue')
    plt.xlabel('Years of Experience',color='blue')
    plt.ylabel("Salary",color='blue')
    plt.title(f'Salary v/s Experience in Degree {i}\nAccuracy Score:- {r2}',color='red')
    plt.show()
    print('---------------------------------------------------------------------------------------------------------------------')
    print('---------------------------------------------------------------------------------------------------------------------')
```

```python
pf=PolynomialFeatures(degree=5)
X2=pf.fit_transform(X)
model_5=LinearRegression()
model_5.fit(X2,y)
y_pr=model_5.predict(X2)
print('Accuracy Score Without adding Polynomial Features is:- ',r2_score(y,y_pred))
print('Accuracy Score With adding Polynomial Features is:- ',r2_score(y,y_pr))
plt.scatter(X,y,color='red')
plt.plot(X,y_pre,color='blue')
plt.xlabel('Years of Experience',color='blue')
plt.ylabel("Salary",color='blue')
plt.title(f'Salary v/s Experience in Degree {5}\nAccuracy Score:- {r2_score(y,y_pr)}',color='red')
plt.show()
```

```python
print(model_5.coef_)
print(model_5.intercept_)
```

```python
experience=[[eval(input())]]
exp=pf.fit_transform(experience)
model_5.predict(exp)
```
