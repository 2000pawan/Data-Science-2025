# Ridge Regression

# Height_Weight

```python
import pandas as pd
from sklearn.linear_model import Ridge,LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('G:/dataset/MLbookdataset/ML/Ml/23. Ridge Regression/train.csv')
df2=pd.read_csv('G:/dataset/MLbookdataset/ML/Ml/23. Ridge Regression/test.csv')
X_train=df1.iloc[:,0:1].values
y_train=df1.iloc[:,-1].values
X_test=df2.iloc[:,0:1].values
y_test=df2.iloc[:,-1].values
```

```python
# Model using Linear Regression
model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_train)
y_pre=model.predict(X_test)
train_score=r2_score(y_train,y_pred)
test_score=r2_score(y_test,y_pre)
plt.scatter(X_train,y_train,color='blue',marker='o')
plt.scatter(X_test,y_test,color='red',marker='d')
plt.title(f'Weight V/s Height\n Training Score:- {train_score} ____ Testing Score:- {test_score}')
plt.show()
plt.scatter(X_train,y_train,color='blue')
plt.plot(X_train,y_pred,color='orange')
plt.show()
```

```python
model=Ridge(alpha=0.001)
model.fit(X_test,y_test)
print(model.score(X_test,y_test))
plt.scatter(X_test,y_test,color='red')
plt.plot(X_test,model.predict(X_test),color='green')
plt.show()
```

```python
print(model.predict([[70.3]]))
```

# Boston Houses

```python
import pandas as pd
from sklearn.linear_model import Ridge,LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('G:/dataset/MLbookdataset/ML/Ml/23. Ridge Regression/boston_houses.csv')
X=df.iloc[:,5:6].values
y=df.iloc[:,-1].values
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1,test_size=0.2)
plt.scatter(X_train,y_train,color='blue',marker='^')
plt.scatter(X_test,y_test,color='red',marker='+')
plt.show()
model=LinearRegression()
model.fit(X_train,y_train)
print('Accuracy Score of model:- ',model.score(X_train,y_train))
plt.scatter(X_train,y_train,color='blue',marker='^')
plt.plot(X_train,model.predict(X_train),color='red')
plt.show()
model.predict([[6.5]])
```

```python
# Training Data
model=Ridge(alpha=0.01)
model.fit(X_train,y_train)
print(model.score(X_train,y_train))
plt.scatter(X_train,y_train,color='red',marker='+')
plt.plot(X_train,model.predict(X_train),color='green')
plt.show()
model.predict([[6.5]])
```

```python
# Testig Data
model=Ridge(alpha=0.01)
model.fit(X_test,y_test)
print(model.score(X_test,y_test))
plt.scatter(X_test,y_test,color='red',marker='+')
plt.plot(X_test,model.predict(X_test),color='green')
plt.show()
model.predict([[6.5]])
```

# Lasso Regression

# Boston Houses

```python
import pandas as pd
from sklearn.linear_model import Lasso,LinearRegression
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('G:/dataset/MLbookdataset/ML/Ml/24. Lasso Regression/boston_houses.csv')
df.drop(columns=['Unnamed: 0'],axis=0,inplace=True)
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
names=list(df.columns)
names.remove('MEDV')
rng=range(X.shape[1])
model=Lasso(alpha=0.1)
model.fit(X,y)
cf=model.coef_
print('Accuracy Score of Model is:- ',model.score(X,y))
plt.plot(rng,cf)
plt.xticks(rng,names,rotation=60)
plt.ylabel("Coefficients")
plt.show()
```

```python
newdata=[[0.02731,0,7.07,0,0.0469,6.421,78.9,4.9671,2,242,17.8,396.90,9.14]] #21.6
model.predict(newdata)
```

# Advertising data

```python
import pandas as pd
from sklearn.linear_model import Lasso
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('G:/dataset/MLbookdataset/ML/Ml/24. Lasso Regression/Advertising.csv')
df.drop(columns=['Unnamed: 0'],axis=0,inplace=True)
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
names=list(df.columns)
names.remove('sales')
rng=range(X.shape[1])
model=Lasso(alpha=0.01)
model.fit(X,y)
cf=model.coef_
print('Accuracy Score of Model is:- ',model.score(X,y))
plt.plot(rng,cf)
plt.xticks(rng,names,rotation=60)
plt.ylabel("Coefficients")
plt.show()
```

# Elasticnet Regression

# Boston Houses 

```python
import pandas as pd
from sklearn.linear_model import ElasticNet
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('G:/dataset/MLbookdataset/ML/Ml/25. ElasticNet Regression/boston_houses.csv')
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
names=list(df.columns)
names.remove('MEDV')
rng=range(X.shape[1])
model=ElasticNet(l1_ratio=0.5)
model.fit(X,y)
cf=model.coef_
c=cf.argmax()
print('Most Influencing column:- ',c,names[c])
print('Accuracy Score of Model is:- ',model.score(X,y))
plt.plot(rng,cf)
plt.xticks(rng,names,rotation=60)
plt.ylabel("Coefficients")
plt.show()
```

```python
model.predict([[0.02731,0,7.07,0,0.0469,6.421,78.9,4.9671,2,242,17.8,396.90,9.14]])
```
