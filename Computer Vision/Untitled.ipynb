```python
import os
from pathlib import Path

folder_path = Path('Songs/')
audio_extensions = {'.mp3', '.mpa', '.wav', '.flac', '.aac', '.ogg', '.wma'}

renamed_count = 0
counter = 1

if not folder_path.exists():
    print(f"Folder not found: {folder_path}")
    exit()

for file_path in folder_path.iterdir():
    if file_path.is_file() and file_path.suffix.lower() in audio_extensions:
        original_name = file_path.stem
        ext = file_path.suffix

        # Step 1 & 2: Take first 5 chars
        new_name = original_name[:5]

        # Step 3: If empty, fallback name
        if not new_name.strip():
            new_name = f"file_{counter:03d}"
            counter += 1

        # Prepare new filename and path
        new_filename = new_name + ext
        new_path = file_path.with_name(new_filename)

        # Handle collisions by appending _1, _2, _3, etc.
        suffix_counter = 1
        base_name = new_name
        while new_path.exists():
            new_filename = f"{base_name}_{suffix_counter}{ext}"
            new_path = file_path.with_name(new_filename)
            suffix_counter += 1

        try:
            file_path.rename(new_path)
            print(f'Renamed: "{file_path.name}" → "{new_filename}"')
            renamed_count += 1
        except Exception as e:
            print(f'❌ Error renaming "{file_path.name}": {e}')

print(f"\n✅ Done: {renamed_count} file(s) renamed.")
```
